### 移动项目到数据盘（只需要运行一次）

```python
%%bash

target_dir="/root/autodl-tmp"
source_dir="facefusion"
input_dir="input"
output_dir="output"

if [ ! -d "${target_dir}/${source_dir}" ]; then
  mv "/root/${source_dir}" "${target_dir}/"
  echo "项目已移动至数据盘"
else
  echo "项目已移动至数据盘"
fi
```
